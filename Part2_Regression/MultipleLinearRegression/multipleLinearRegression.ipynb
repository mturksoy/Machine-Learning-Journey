MULTIPLE LINEAR REGRESSION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('50_Startups.csv')
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print(X)
```

```python
print(y)
```

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= 0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2) #virgülden sonra sadece 2 basamağı dahil eder
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)), 1)) 
#sonuçları dikey olarak görebilmek için resahpe ile dikey hale getirdik
#yan yana gözükmesi için de axis değerini 1 verdik, 0 verirsek alt alta olur.
```

```python
print(regressor.predict([[1, 0, 0, 160000, 130000, 300000]])) # predicting just one data
```

```python
print(regressor.coef_)   
print(regressor.intercept_)
```
